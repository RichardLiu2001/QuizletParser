```python
def parseRow(traditional, simplified, pinyin, english):

    traditionalCard = str(traditional) + "|" + str(pinyin) + " - " + str(english)
    simplifiedCard = ""
    if(str(simplified) != 'nan'):
        traditionalCard += " (Traditional)"
        simplifiedCard = str(simplified) + "|" + str(pinyin) + " - " + str(english)
        simplifiedCard=simplifiedCard.replace('\n',' ')
        print(simplifiedCard)
    traditionalCard = traditionalCard.replace('\n', ' ')
    print(traditionalCard)
```

```python
first_unit.apply(lambda x: parseRow(x['Traditional'], x['Simplified'], x['Pinyin'], x['English']), axis=1)

if (second_start != 0):
    second_unit.apply(lambda x: parseRow(x['Traditional'], x['Simplified'], x['Pinyin'], x['English']), axis=1)
```
